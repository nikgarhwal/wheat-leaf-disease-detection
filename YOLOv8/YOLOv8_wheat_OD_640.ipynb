```python
#import the YOLOv8 model
from ultralytics import YOLO
model = YOLO('yolov8m.yaml')  # build a new model from YAML
model = YOLO('yolov8m.pt')  # load a pretrained model (recommended for training)
model = YOLO('yolov8m.yaml').load('yolov8m.pt')
```

```python
#train the model
model.train(data='/content/drive/MyDrive/dissertation_project/wheat_dataset/wheat_dataset_640x640/yolov8_wheat.yaml', epochs=1000, imgsz=640, batch= 16, device='1', project='/content/drive/MyDrive/dissertation_project/model_training/yolov8', plots=True, resume=True)
```

```python
# Load a model
model = YOLO('/content/drive/MyDrive/dissertation_project/model_training/yolov8/train5/weights/best.pt')  # load a custom model
```

```python
# Validate the model
metrics = model.val()  # no arguments needed, dataset and settings remembered
metrics.box.map    # map50-95
metrics.box.map50  # map50
metrics.box.map75  # map75
metrics.box.maps   # a list contains map50-95 of each category
```
