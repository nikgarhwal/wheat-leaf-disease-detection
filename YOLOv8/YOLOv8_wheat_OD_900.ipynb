```python
#import the YOLOv8 model
from ultralytics import YOLO
model = YOLO('yolov8m.yaml').load('yolov8m.pt')
```

```python
#train the model
model.train(data='/content/drive/MyDrive/YOLOv8/plantData.yaml', epochs=150, imgsz=900, batch= 8, device='1', project='/content/drive/MyDrive/YOLOv8/training_results', plots=True, resume=True)
```

```python
model = YOLO('/content/drive/MyDrive/YOLOv8/training_results/train4/weights/best.pt')
```

```python
model.val()
```

```python
# Run batched inference on a list of images
results = model('/content/drive/MyDrive/YOLOv8/DataSet/images/test', conf=0.45, save=True,  save_conf=True)  # return a list of Results objects
```
